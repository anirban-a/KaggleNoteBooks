```r
library(tidyverse) # metapackage of all tidyverse packages
library(dplyr)
library(ggplot2)

# Input data files are available in the read-only "../input/" directory

# Link: https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/overview
```

```r
# Read file
df <- read.csv('../input/house-prices-advanced-regression-techniques/train.csv')
```

## Exploring the dataset

```r
ncol(df)
nrow(df)
```

## Explore variables for correlation with SalePrice

```r
ggplot(df, aes(SalePrice, TotalBsmtSF))+
    geom_point()
```

```r
# view the distribution of TotalBsmtSF
ggplot(df, aes(x=TotalBsmtSF))+geom_histogram(stat = 'density')

ggplot(df, aes(x=log(TotalBsmtSF)))+geom_histogram(stat = 'density')
```

```r
# log normalize `TotalBsmtSF`
df$TotalBsmtSF <- log(df$TotalBsmtSF)
```

```r
ggplot(df, aes(SalePrice, TotalBsmtSF))+
    geom_point()
```

## Inspect how LotShape is affecting the SalePrice

```r
ggplot(df, aes(x=SalePrice, color=as.factor(LotShape)))+
    theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) +
    geom_histogram(bins=50)+facet_grid(~LotShape)
```

### Looks like plot shape has direct correlation with the SalePrice.

more number of plots with regular shape is sold at a lower price. As shape gets more and more irregular, the sales drop
